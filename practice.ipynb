```python
x=np.linspace(0,10,100)
x1=np.sin(x)
y1=np.cos(x)
plt.style.use("ggplot")
plt.figure(figsize=(12,4))
plt.plot(x,x1,label="sinx")
plt.plot(x,y1, label = "cosx")

plt.title("cosine and sin curve",size=20)
plt.xlabel("x-axis",color="red")
plt.ylabel("y-axis",color="blue")
plt.legend(title="trigo fxn",fontsize="large",loc ="upper right",edgecolor="green")
plt.show() 

```

```python
import pandas as pd

data1 = {
    'A': [1, 2, 3],
    'B': [4, 5, 6]
}
data2 = {
    'A': [7, 8, 9],
    'B': [10, 11, 12]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Concatenate DataFrames vertically (default)
df_vertical = pd.concat([df1, df2], ignore_index=True)
print(df_vertical)

# Concatenate DataFrames horizontally
```

```python
df_horizontal = pd.concat([df1, df2], axis=1)
print(df_horizontal)
```

```python
df_joined=df1.join(df2,how="left",lsuffix='_left')
df_joined
```

```python
import numpy as np
from sklearn.impute import KNNImputer
data = np.array([[1, 2, np.nan],
                 [3, 4, 3],
                 [np.nan, 6, 5],
                 [8, 8, 7]])
imputer = KNNImputer(n_neighbors=2)
imputed_data = imputer.fit_transform(data)

print(imputed_data)
```
