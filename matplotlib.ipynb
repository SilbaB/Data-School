```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label="sinx",color="violet",linewidth=3)
plt.plot(x,y2,color="black",label="cosx",linewidth=0.8)
plt.grid(color="cyan", linestyle="--")
plt.legend(loc='upper right')
plt.show()
plt.savefig("sin_cos_functions.png")
```

*Customized plot*

```python
x=np.linspace(0,10,100)
x1=np.sin(x)
y1=np.cos(x)
plt.figure(figsize=(7,5))
plt.plot(x,x1,label='sinx',linestyle='--',marker='*')
plt.plot(x,y1,label='cosx',linestyle='-',marker='s')
plt.legend(title="trig fxn",loc='upper right',fontsize=9,edgecolor='cyan')
plt.show()
```

```python

# Create a simple plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

# Add text and annotation
plt.text(5, 0, 'Text at (5, 0)')
plt.annotate('Local max', xy=(1.57, 1), xytext=(2, 0.7),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Show the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Text and Annotation Example')
plt.grid(True)
plt.show()
```
