```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label="sinx",color="yellow",linewidth=0.8)
plt.plot(x,y2,color="black",label="cosx",linewidth=0.8)
plt.grid(color="gray", linestyle="--")
plt.legend()
plt.show()
plt.savefig("sin_cos_functions.png")
```
